```python
import os

# Path to the images directory
images_directory = '../images/'

# Getting the list of image files in the directory
image_files = os.listdir(images_directory)

# Creating the LaTeX code for each image
latex_code = ""
for image in image_files:
    latex_code += f"\\begin{{figure}}[ht]\n"
    latex_code += "    \\centering\n"
    latex_code += f"    \\includegraphics[width=1.0\\textwidth]{{images/{image}}}\n"
    latex_code += "    \\caption{Caption for the image}\n"
    latex_code += f"    \\label{{fig:{image.split('.')[0]}}}\n"
    latex_code += "\\end{figure}\n\n"

print(latex_code)
```
